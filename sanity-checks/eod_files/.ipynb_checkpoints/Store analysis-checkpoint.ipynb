<h3>This is the analysis for a store based on the historical data of the releases</h3>

```python
import numpy as np 
import pandas as pd
```

Data Preparation 

```python
data = pd.read_csv('southampton.csv')
```

```python
data.head(5)
```

```python
data_dict = data.to_dict()
```

```python
data_dict.keys()
```

```python
del data_dict[' ST File Size']
del data_dict[' PL File Size']
del data_dict[' Replication Window']
del data_dict['Name']
del data_dict['EOD Start']
del data_dict['EOD Finish']
del data_dict['Country']
del data_dict[' Backup Start']
del data_dict[' Backup End']
del data_dict[' Replication Start']
del data_dict[' Replication End']
del data_dict[' Deployment Start']
del data_dict[' Deployment End']
del data_dict['Store']
```

```python
data_dict.keys()
```

```python
data = pd.DataFrame(data_dict)
```

```python
data.columns
```

```python
data = data.rename(columns={' Dep Duration' : 'Dep_Duration',  
                     ' Replication Duration(s)' : 'Replication_Duration', 
                     'EOD Duration(s)' : 'EOD_Duration'})
```

```python
data = data.fillna(0)
```

```python
data = data.replace('-',0)
```

```python
data
```

```python
#data.info()
data['EOD_Duration'] = data['EOD_Duration'].astype(float)
data['Replication_Duration'] = data['Replication_Duration'].astype(float)
data['Dep_Duration'] = data['Dep_Duration'].astype(float)
```

<h3> exploratory analysis </h3>

maximum time taken for an End of Day process

```python
maxEOD = data[['Date','EOD_Duration']].where(data['EOD_Duration'] == data['EOD_Duration'].max())
maxEOD.dropna()
```

minmum time taken for an End of Day process

```python
data['EOD_Duration'] = data['EOD_Duration'].replace(0,np.nan)
minEOD = data[['Date','EOD_Duration']].where(data['EOD_Duration'] == data['EOD_Duration'].min())
minEOD.dropna()
```

scatter plot

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
x = data['Date']
y = data['EOD_Duration'].replace(0,np.nan)
y.unique()
plt.plot(x,y,alpha = 0.5)
plt.show()
```

```python
data['EOD_Duration'].describe()
```

```python
print(data['EOD_Duration'].isnull().sum(axis = 0),'Times EOD process failed for this store')
```

```python
print(data['EOD_Duration'].count(),'Times EOD process was successful for this store')
```

```python
y = data['EOD_Duration'].isnull().sum(axis = 0)
x = data['EOD_Duration'].count()
labels = 'Successful EOD', 'Failed EOD'
sizes = [x,y]
colors = ['Green', 'Red']
explode = (0.1, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```
